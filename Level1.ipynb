```python
#Barplot to visualize distribution of prize range among restaurants
plt.figure(figsize=(10, 6))
df['Price range'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Price Ranges among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

**Task 4**

Task: Online Delivery
Determine the percentage of restaurants
that offer online delivery.
Compare the average ratings of restaurants
with and without online delivery.

```python
#percentage of restaurantss with online delivery
df['Has Online delivery'] = df['Has Online delivery'].astype(str).str.lower() == 'yes'
restaurants_with_online_delivery = df['Has Online delivery'].sum()
total=len(df)
percentage= (restaurants_with_online_delivery/total)*100
percentage
```

Approx 25% restaurants offer the faacility of online delivery

```python
#Compairing the rating of restaurants with and without onlive delivery
avg1= df[df['Has Online delivery']]['Aggregate rating'].mean()
avg2 = df[~df['Has Online delivery']]['Aggregate rating'].mean()
print("Average rating of restaurants with onlive delivery is : ", avg1)
print("Average rating of restaurants without online delivery is : ", avg2)
```
