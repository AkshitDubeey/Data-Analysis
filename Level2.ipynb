Task: Cuisine Combination Identify the most common combinations of cuisines in the dataset. Determine if certain cuisine combinations tend to have higher ratings.

```python
#Identifing the most common combinations of cuisines in the  dataset
df['Cuisines'] = df['Cuisines'].astype(str)
df['Cuisines'] = df['Cuisines'].str.split(', ')
df['Cuisine Combination'] = df['Cuisines'].apply(lambda x: ', '.join(sorted(x))) #creating a new collumn list of cuisines combinatioins
cuisine_count = df['Cuisine Combination'].explode().value_counts() #counting occurence of each combination
top=cuisine_count.head(5)
top
```

```python
#Determining if certain cuisine combinations tend to have higher ratings
avg = df.groupby('Cuisine Combination')['Aggregate rating'].mean()#Calculating the average rating for each cuisine combination
#Combining  the counts and average ratings
cuisine_combinations_analysis = pd.DataFrame({
    'Count': cuisine_count,
    'Average Rating': avg
}).reset_index()
#Combinations with higher average ratings are:
print(cuisine_combinations_analysis.sort_values(by='Average Rating', ascending=False).head(10))
```

#Task 2 Observations
1. The most common combination of cuisines in the dataset is "Chinese, North Indian"
2. "Bakery, Continental, Italian" This Cuisine combination tend to have heighter ratings.

Task: Geographic Analysis Plot the locations of restaurants on a map using longitude and latitude coordinates. Identify any patterns or clusters of restaurants in specific areas

```python
import folium #For creating an interactive map
from folium.plugins import MarkerCluster
```

```python
#Creating a map centered at speccific location
map_center=[28.64746, 77.12007]
my_map= folium.Map(location=map_center,zoom_start=12)
#Now we have to add markers for each restaurant
marker_cluster = MarkerCluster().add_to(my_map)
marker=MarkerCluster().add_to(my_map)
#now giving coordinates
for index, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name'], 
        icon=folium.Icon(color='red'), 
    ).add_to(marker_cluster)
my_map.save("restaurant_map.html")
```

Task: Restaurant Chains Identify if there are any restaurant chains present in the dataset. Analyze the ratings and popularity of different restaurant chains.

```python
#Identifing restaurnt chains 
count= df["Restaurant Name"].value_counts() #Counting restaurants of evvery restaurant chain
chains=count[count>1]
chains
```

```python
#Let's create a new dataframe with only the restaurant chains.
df_chains= df[df["Restaurant Name"].isin(chains.index)]
df_chains
```

```python
#Now we will analyze the ratings of different chains.
ratings= df_chains.groupby("Restaurant Name")["Aggregate rating"].mean().sort_values(ascending=False)
ratings
```

```python
#analysing popularity
votes= df_chains.groupby("Restaurant Name")["Votes"].sum().sort_values(ascending=False)
votes
```

Through the nummber of votes we can analyze the popularity of different restaurant chains. according to which Barbeque Nation is one of the most popular restaurant chain.
